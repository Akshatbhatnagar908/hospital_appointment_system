```python
import pandas as pd
import os


def add_patient():
    
        name = input("\nEnter Patient Name: ")
        age = input("Enter Patient Age: ")
        gender = input("Enter Patient Gender: ")
        disease = input("Enter Disease: ")
        contact = input("Enter Contact Number: ")

        # Validate numeric fields
        if not age.isdigit():
            print("Invalid age. Please enter a number.")
           
        if not contact.isdigit():
            print("Invalid contact number. Please enter digits only.")
           

        # Create a dictionary for patient data
        patient_data = {
            "Name": [name],
            "Age": [int(age)],
            "Gender": [gender],
            "Disease": [disease],
            "Contact": [contact]
        }

        # Convert dictionary to DataFrame
        df = pd.DataFrame(patient_data)

        # Check if file exists and append data
        if os.path.exists(file_name):
            existing_df = pd.read_excel(file_name)
            df = pd.concat([existing_df, df], ignore_index=True)

        # Save to Excel file
        df.to_excel(file_name, index=False)
        print(f"Patient record saved successfully in {file_name}.")

# Run the function
add_patient()
```

```python
import pandas as pd
import tkinter as tk
from tkinter import ttk, messagebox
import time
from datetime import datetime

# Load the patient data from the Excel file
def load_patient_data():
    try:
        # Read the Excel file (replace 'patients.xlsx' with the correct file path)
        df = pd.read_excel('patients.xlsx', engine='openpyxl')
        return df
    except Exception as e:
        print(f"Error reading Excel file: {e}")
        return None

# Global variables to track start and end times
appointments = {}

# Function to start an appointment for the selected patient
def start_appointment():
    selected_item = tree.selection()
    if selected_item:
        patient_data = tree.item(selected_item[0])['values']
        patient_name = patient_data[0]
        
        # Record the start time when the appointment is started
        start_time = datetime.now()
        appointments[patient_name] = {'start_time': start_time, 'end_time': None}
        
        # Display start appointment info
        messagebox.showinfo("Appointment Status", f"Appointment started for {patient_name} at {start_time.strftime('%H:%M:%S')}")
    else:
        messagebox.showwarning("No Patient Selected", "Please select a patient to start the appointment.")

# Function to stop an appointment for the selected patient
def stop_appointment():
    selected_item = tree.selection()
    if selected_item:
        patient_data = tree.item(selected_item[0])['values']
        patient_name = patient_data[0]
        
        # Record the end time when the appointment is stopped
        if patient_name in appointments and appointments[patient_name]['start_time']:
            end_time = datetime.now()
            appointments[patient_name]['end_time'] = end_time
            
            # Calculate the turnaround time (time spent in appointment)
            turnaround_time = end_time - appointments[patient_name]['start_time']
            total_time_taken = sum([(appointment['end_time'] - appointment['start_time']).total_seconds() for appointment in appointments.values() if appointment['end_time']])

            # Show turnaround time for the specific patient
            messagebox.showinfo("Appointment Status", f"Appointment stopped for {patient_name} at {end_time.strftime('%H:%M:%S')}\nTurnaround Time: {turnaround_time}")
            
            # Update the total time taken
            total_time_label.config(text=f"Total Time Taken: {total_time_taken // 3600} hours, {(total_time_taken % 3600) // 60} minutes, {total_time_taken % 60} seconds")
        else:
            messagebox.showwarning("Error", f"No appointment has been started for {patient_name}.")
    else:
        messagebox.showwarning("No Patient Selected", "Please select a patient to stop the appointment.")

# Function to display patient details in a Tkinter window
def show_patient_details(df):
    if df is not None:
        window = tk.Tk()
        window.title("Patient Details")

        # Create a treeview widget to display the data
        global tree
        tree = ttk.Treeview(window, columns=("Name", "Age", "Disease", "Contact", "Gender"), show="headings")
        
        # Define the column headings
        tree.heading("Name", text="Name")
        tree.heading("Age", text="Age")
        tree.heading("Disease", text="Disease")
        tree.heading("Contact", text="Contact")
        tree.heading("Gender", text="Gender")
        
        # Insert the patient data into the treeview
        for _, row in df.iterrows():
            tree.insert("", tk.END, values=(row["Name"], row["Age"], row["Disease"], row["Contact"], row["Gender"]))
        
        # Create Start and Stop Appointment buttons
        start_button = tk.Button(window, text="Start Appointment", command=start_appointment)
        start_button.pack(pady=10)
        
        stop_button = tk.Button(window, text="Stop Appointment", command=stop_appointment)
        stop_button.pack(pady=10)
        
        # Label to show total time taken
        global total_time_label
        total_time_label = tk.Label(window, text="Total Time Taken: 0 hours, 0 minutes, 0 seconds")
        total_time_label.pack(pady=20)

        # Pack the treeview into the window and run the application
        tree.pack(fill=tk.BOTH, expand=True)
        window.mainloop()

if __name__ == "__main__":
    df = load_patient_data()
    show_patient_details(df)

```
